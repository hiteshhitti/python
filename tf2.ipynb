```python
# ✅ Import Required Libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

# ✅ Step 1: Prepare Dummy Input Data (X) and Output Labels (y)
# Example: XOR Problem Data
X = np.array([
    [0, 0],








    
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 0])  # Output for XOR problem

# ✅ Step 2: Build ANN Model using Sequential API
model = Sequential()

# Add Input Layer + Hidden Layer
model.add(Dense(4, activation='relu', input_dim=2))  # 4 Neurons, ReLU Activation

# Add Output Layer
model.add(Dense(1, activation='sigmoid'))  # Binary output, so Sigmoid

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# ✅ Step 4: Train the Model
model.fit(X, y, epochs=100, verbose=1)  # 100 epochs, show output progress

# ✅ Step 5: Evaluate the Model (Check Accuracy)
loss, accuracy = model.evaluate(X, y)
print(f"\nFinal Accuracy on Training Data: {accuracy*100:.2f}%")

# ✅ Step 6: Make Predictions
predictions = model.predict(X)
print("\nModel Predictions (Raw Output):")
print(predictions)

# ✅ Step 7: Convert Probabilities to Binary Output (0 or 1)
predicted_classes = (predictions > 0.5).astype(int)
print("\nPredicted Classes (After Thresholding at 0.5):")
print(predicted_classes)

# ✅ Optional: Display Actual vs Predicted
print("\nActual Output:", y)
print("Predicted Output:", predicted_classes.flatten())
```
