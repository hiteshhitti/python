```python
top_states = honey.groupby('state')['production'].sum().sort_values(ascending=False).head(5).index.tolist()

top_honey = honey[honey['state'].isin(top_states)]

g = sns.FacetGrid(top_honey, col="state", col_wrap=3, height=3)
g.map(sns.lineplot, "year", "production")
plt.show()
```

```python
# Q1, Q2, Q3 using pandas quantile()
Q1 = honey['yield_per_colony'].quantile(0.25)
Q2 = honey['yield_per_colony'].quantile(0.50)
Q3 = honey['yield_per_colony'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Fences for Outlier detection
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

print("Q1 (25th Percentile):", Q1)
print("Q2 (Median):", Q2)
print("Q3 (75th Percentile):", Q3)
print("IQR:", IQR)
print("Lower Fence:", lower_fence)
print("Upper Fence:", upper_fence)

# Detect outliers
outliers = honey[(honey['yield_per_colony'] < lower_fence) | (honey['yield_per_colony'] > upper_fence)]
print("\nOutlier Records:\n", outliers)
clean_honey = honey[(honey['yield_per_colony'] >= lower_fence) & (honey['yield_per_colony'] <= upper_fence)]
print(clean_honey)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=clean_honey['yield_per_colony'])
plt.title('Boxplot showing IQR and Outliers (Yield per Colony)')
plt.show()
```
