```python
# ============================================================
# Stop-word Removal + Text Cleaning Pipeline
# ============================================================

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import re

stop_words = set(stopwords.words('english'))
print(f"Total English stop words: {len(stop_words)}")
print(f"Sample: {list(stop_words)[:10]}")

def clean_text(text):
    """Complete text cleaning pipeline."""
    # Step 1: Lowercase
    text = text.lower()
    # Step 2: Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # Step 3: Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    # Step 4: Remove numbers
    text = re.sub(r'\d+', '', text)
    # Step 5: Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Step 6: Tokenize
    tokens = word_tokenize(text)
    # Step 7: Remove stop words and short tokens
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    return tokens

raw = "Send me an email at john@test.com or visit https://nlp.org for the 3 best NLP techniques!"
cleaned = clean_text(raw)

print(f"\nOriginal : {raw}")
print(f"Cleaned  : {cleaned}")
print(f"Reduction: {len(raw.split())} words → {len(cleaned)} tokens")
```

```python
# ============================================================
# Stemming vs Lemmatization — Key Difference Demo
# ============================================================

from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
from nltk.stem import WordNetLemmatizer

porter  = PorterStemmer()
snowball = SnowballStemmer("english")
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

words = ['running', 'runs', 'ran', 'easily', 'fairly',
         'caring', 'studies', 'studied', 'better', 'corpora']

print(f"{'Word':<12} {'Porter':<12} {'Snowball':<12} {'Lancaster':<12} {'Lemma(v)':<12} {'Lemma(n)'}")
print("-" * 75)
for word in words:
    print(f"{word:<12}"
          f"{porter.stem(word):<12}"
          f"{snowball.stem(word):<12}"
          f"{lancaster.stem(word):<12}"
          f"{lemmatizer.lemmatize(word, pos='v'):<12}"
          f"{lemmatizer.lemmatize(word, pos='n')}")

print("\n💡 Key Insight:")
print("  'better' → Porter: 'better' | Lemma(adj): 'good'")
print("  'corpora' → Porter: 'corpora' | Lemma(n): 'corpus'")
print("  Lemmatization is linguistically correct but needs POS tag.")
```

```python
# ============================================================
# HOUR 4: Sentiment Analysis
# ============================================================

# pip install textblob vaderSentiment
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
import nltk
nltk.download('vader_lexicon')

vader = SentimentIntensityAnalyzer()

reviews = [
    "This movie is absolutely fantastic! Best film ever made.",
    "Terrible waste of time. I hated every minute of it.",
    "It was okay, nothing special but not bad either.",
    "NOT bad at all! Surprisingly good despite low budget.",
    "The acting was good but the plot was boring and predictable.",
    "😍 Amazing!! Loved it so much I watched it 3 times!",
]

print(f"{'Review':<50} {'VADER':<10} {'TextBlob':<10} {'Label'}")
print("-" * 85)

for review in reviews:
    # VADER scores
    vs = vader.polarity_scores(review)
    compound = vs['compound']

    # TextBlob scores
    blob = TextBlob(review)
    tb_polarity = blob.sentiment.polarity

    # Label from VADER compound score
    if compound >= 0.05:
        label = "POSITIVE"
    elif compound <= -0.05:
        label = "NEGATIVE"
    else:
        label = "NEUTRAL"

    short = review[:48] + ".." if len(review) > 48 else review
    print(f"{short:<50} {compound:>+.3f}    {tb_polarity:>+.3f}     {label}")

# Detailed breakdown for one review
print("\n=== Detailed VADER breakdown ===")
scores = vader.polarity_scores(reviews[0])
print(f"Positive: {scores['pos']:.3f}")
print(f"Negative: {scores['neg']:.3f}")
print(f"Neutral : {scores['neu']:.3f}")
print(f"Compound: {scores['compound']:.3f}  ← main score [-1, +1]")
```

```python
# ============================================================
# Named Entity Recognition (NER) with spaCy
# ============================================================

import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")

text = """Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne
on April 1, 1976, in Cupertino, California. In 2023, Apple became the
first company to surpass $3 trillion in market capitalization. Tim Cook,
who took over as CEO in 2011, announced the new iPhone 15 at Apple Park.
The company collaborates with partners in China, India, and South Korea."""

doc = nlp(text)

# Print all entities
print("=== Named Entities Found ===")
print(f"{'Entity':<25} {'Label':<12} {'Description'}")
print("-" * 65)
for ent in doc.ents:
    desc = spacy.explain(ent.label_)
    print(f"{ent.text:<25} {ent.label_:<12} {desc}")

# Count entity types
print("\n=== Entity Distribution ===")
ent_counts = Counter([ent.label_ for ent in doc.ents])
for label, count in ent_counts.most_common():
    bar = "█" * (count * 5)
    print(f"  {label:<10} {bar} ({count})")

# Custom NER with PhraseMatcher
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab, attr="LOWER")
custom_terms = [nlp("machine learning"), nlp("deep learning"), nlp("neural network")]
matcher.add("AI_TERMS", custom_terms)

doc2 = nlp("Machine learning and deep learning power many AI systems. Neural networks are key.")
matches = matcher(doc2)
print("\n=== Custom Matcher Results ===")
for match_id, start, end in matches:
    print(f"  Found: '{doc2[start:end]}' at positions [{start}:{end}]")
```

```python
# ============================================================
# HOUR 5: Spam SMS Classifier — Full Pipeline
# Dataset: https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection
# Or use: pip install ucimlrepo
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re, string

# --- Load dataset ---
# Option A: from file
df = pd.read_csv('spam.csv',encoding='latin-1' ,sep='\t',
                 header=None, names=['label', 'message'])

# Option B: create small sample for classroom
data = {
    'label': ['ham','spam','ham','spam','ham','spam','ham','spam','ham','spam',
              'ham','ham','spam','ham','spam'],
    'message': [
        "Hey, are you coming to the meeting tomorrow?",
        "FREE! Win £1000 cash! Call now 08000930705 to claim!",
        "Can you pick up milk on your way home?",
        "URGENT: Your mobile No. awarded a £2,000 prize! Reply WIN!",
        "I'll be there in 10 minutes, stuck in traffic.",
        "Congratulations! You've won a free iPhone. Click to claim!",
        "Let's catch up for coffee this weekend",
        "Get a £500 voucher NOW! Limited time offer. Text CLAIM to 81010",
        "Happy birthday! Hope you have a wonderful day",
        "You have won a SIM card with unlimited credit! Call 09061743386",
        "Are you free for dinner on Saturday?",
        "Don't forget mum's birthday is next week!",
        "WINNER!! You have been selected for a cash prize. Text NOW!",
        "The project report is due Friday, please review it.",
        "Earn £800/week from home! No experience needed. Text START",
    ]
}
df = pd.DataFrame(data)

# --- Explore ---
print("=== Dataset Info ===")
print(df.info())
print("\n=== Class Distribution ===")
print(df['label'].value_counts())
print(f"\nSpam ratio: {df[df.label=='spam'].shape[0]/len(df):.1%}")

# --- Feature engineering ---
df['msg_length'] = df['message'].apply(len)
df['word_count'] = df['message'].apply(lambda x: len(x.split()))
df['has_number'] = df['message'].apply(lambda x: bool(re.search(r'\d', x)))
df['caps_ratio'] = df['message'].apply(
    lambda x: sum(1 for c in x if c.isupper()) / len(x) if len(x) > 0 else 0
)

print("\n=== Feature Stats by Label ===")
print(df.groupby('label')[['msg_length', 'word_count', 'caps_ratio']].mean().round(2))
```

```python
# ============================================================
# Part 2: Preprocessing + Naive Bayes Training
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    """Full preprocessing pipeline for spam classification."""
    text = text.lower()
    text = re.sub(r'\d+', 'NUM', text)   # normalize numbers
    text = re.sub(r'http\S+', 'URL', text) # normalize URLs
    text = re.sub(r'[^\w\s]', '', text)   # remove punctuation
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(t) for t in tokens
              if t not in stop_words and len(t) > 1]
    return ' '.join(tokens)

# Apply preprocessing
df['clean_msg'] = df['message'].apply(preprocess)
print("=== Preprocessing Examples ===")
for _, row in df.head(3).iterrows():
    print(f"  Original : {row['message']}")
    print(f"  Cleaned  : {row['clean_msg']}")
    print()

# Encode labels
le = LabelEncoder()
df['label_enc'] = le.fit_transform(df['label'])  # ham=0, spam=1

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_msg'], df['label_enc'],
    test_size=0.2, random_state=42, stratify=df['label_enc']
)
print(f"Train size: {len(X_train)} | Test size: {len(X_test)}")

# Build Pipeline: TF-IDF → Naive Bayes
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        ngram_range=(1,2),
        min_df=1,
        max_features=5000,
        sublinear_tf=True
    )),
    ('clf', MultinomialNB(alpha=0.1))
])

pipeline.fit(X_train, y_train)
print("Model trained successfully!")
```

```python
# ============================================================
# Part 4: Interactive Prediction + Cross-Validation
# ============================================================

from sklearn.model_selection import cross_val_score, StratifiedKFold

def predict_message(message, threshold=0.5):
    """Predict spam with detailed explanation."""
    cleaned = preprocess(message)
    prob = pipeline.predict_proba([cleaned])[0]
    ham_prob, spam_prob = prob[0], prob[1]
    label = "SPAM" if spam_prob >= threshold else "HAM"

    print(f"\n{'='*55}")
    print(f"Message : {message[:60]}..." if len(message)>60 else f"Message : {message}")
    print(f"Cleaned : {cleaned}")
    print(f"{'─'*55}")
    print(f"HAM  probability : {ham_prob:.4f} {'█'*int(ham_prob*30)}")
    print(f"SPAM probability : {spam_prob:.4f} {'█'*int(spam_prob*30)}")
    print(f"PREDICTION       : {'🚫 ' + label if label=='SPAM' else '✅ ' + label}")
    print(f"{'='*55}")
    return label

# Test on new messages
test_messages = [
    "Are you joining us for lunch today?",
    "FREE entry to WIN the weekly competition! Text WIN to 87121",
    "You have been selected for an exclusive offer. Call now!",
    "The exam results will be announced tomorrow at 10am.",
    "Claim your £1000 cash prize TODAY! Limited time only!",
]

for msg in test_messages:
    predict_message(msg)

# --- Cross-validation for reliable accuracy ---
print("\n=== 5-Fold Cross-Validation ===")
X_all = df['clean_msg']
y_all = df['label_enc']
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(pipeline, X_all, y_all, cv=skf, scoring='f1')
print(f"F1 scores per fold : {np.round(cv_scores, 4)}")
print(f"Mean F1            : {cv_scores.mean():.4f}")
print(f"Std deviation      : {cv_scores.std():.4f}")
```

```python
# ============================================================
# Bonus: Compare Naive Bayes vs Other Classifiers
# ============================================================

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(ngram_range=(1,2), max_features=5000, sublinear_tf=True)
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec  = tfidf.transform(X_test)

classifiers = {
    'Multinomial NB'    : MultinomialNB(alpha=0.1),
    'Complement NB'     : ComplementNB(alpha=0.1),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Linear SVM'        : LinearSVC(random_state=42, max_iter=2000),
    'Random Forest'     : RandomForestClassifier(n_estimators=100, random_state=42),
}

results = []
print(f"{'Classifier':<22} {'Accuracy':>10} {'F1 (spam)':>12} {'Time':>8}")
print("-" * 55)

import time
for name, clf in classifiers.items():
    t0 = time.time()
    clf.fit(X_train_vec, y_train)
    ypred = clf.predict(X_test_vec)
    elapsed = time.time() - t0
    acc = accuracy_score(y_test, ypred)
    f1  = f1_score(y_test, ypred, pos_label=1, zero_division=0)
    results.append({'clf':name, 'acc':acc, 'f1':f1})
    print(f"{name:<22} {acc:>10.4f} {f1:>12.4f} {elapsed:>6.3f}s")

best = max(results, key=lambda r: r['f1'])
print(f"\nBest model by F1: {best['clf']} (F1={best['f1']:.4f})")
```
