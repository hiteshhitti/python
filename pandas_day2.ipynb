```python
import pandas as pd
```

```python
data = {
    "Name": ["Aman", "Riya", "Karan", "Simran", "Rahul"],
    "Age": [22, 23, 21, 24, 22],
    "City": ["Delhi", "Mumbai", "Delhi", "Chandigarh", "Mumbai"],
    "Marks": [85, 90, 78, 88, 92]
}
```

```python
df=pd.DataFrame(data)
```

```python
df.loc[0]         
```

```python
df.loc[0, "Name"]        # specific value
```

```python
df.loc[0:2]              # inclusive slicing
```

```python
df.loc[:, "Name"]        # only column
```

```python
df.loc[1:3, ["Name","Marks"]]
```

```python
df[df["Age"] > 22]
```

```python
df[df["City"] == "Delhi"]
```

```python
df[df["Marks"] >= 90]
```

```python
df[(df["Age"] > 21) & (df["City"] == "Delhi")]
```

```python
df[(df["City"] == "Delhi") | (df["Marks"] > 90)]
```

```python
df[~(df["City"] == "Delhi")]
```

```python
df.sort_values("Marks")
```

```python
df.sort_values("Marks", ascending=False)
```

```python
df.sort_values(["City", "Marks"])
```

```python
df.sort_values(["City", "Marks"], ascending=[True, False])
```

```python
df.sort_values("Marks", inplace=True)
```
