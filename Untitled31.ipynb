```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
honey=pd.read_csv('US_honey_dataset_updated.csv')
```

```python
alabama=honey[honey['state']=='Alabama']
```

```python
alabama=alabama[['production', 'year']]
```

```python
plt.plot(alabama['year'], alabama['production'])
plt.title('Honey Production in Alabama Over Years')
plt.xlabel('Year')
plt.ylabel('Production')
plt.show()
```

```python
state_prod = honey.groupby('state')['production'].sum().sort_values(ascending=False)
```

```python
state_prod.plot(kind='bar', figsize=(10,6))
plt.title('Total Honey Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.show()
```

```python
plt.figure(figsize=(10,6))
state_prod.plot(kind='bar', color='orange')  # ✅ All bars orange
plt.title('Total Honey Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.show()
```

```python
state_prod.plot(kind='barh', figsize=(10,10))
plt.title('Total Honey Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.show()
```

```python
colors = ['red', 'green', 'blue', 'purple', 'orange', 'cyan', 'magenta', 'yellow'] * 10  # Extend if more states

plt.figure(figsize=(10,6))
state_prod.plot(kind='bar', color=colors[:len(state_prod)])  # ✅ Different color for each bar
plt.title('Total Honey Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.show()
```

```python
cmap = plt.get_cmap('viridis')  # Try others: 'plasma', 'inferno', 'cool', 'spring'

colors = [cmap(i / len(state_prod)) for i in range(len(state_prod))]
```

```python
plt.figure(figsize=(10,6))
state_prod.plot(kind='bar', color=colors)
plt.title('Total Honey Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.show()
```
