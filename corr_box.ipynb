```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("US_honey_dataset_updated.csv")

df.head()
df.info()
df.describe()
```

```python
sns.histplot(df['production'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='state', y='production', data=df)
sns.stripplot(x='state', y='production', data=df,
              color='red', alpha=0.4, jitter=True)
plt.xticks(rotation=90)
plt.show()
```

```python
top_states = df.groupby('state')['production'].sum().nlargest(10).index

filtered_df = df[df['state'].isin(top_states)]

plt.figure(figsize=(10,6))

sns.boxplot(x='state', y='production', data=filtered_df)

sns.stripplot(x='state', y='production', data=filtered_df, 
              color='red', alpha=0.4, jitter=True)

plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(y='production', data=df)
sns.stripplot(y='production', data=df, color='red', alpha=0.3)
plt.show()
```

```python
top_states = df.groupby('state')['production'].sum().nlargest(5).index
filtered_df = df[df['state'].isin(top_states)]

sns.boxplot(x='state', y='production', data=filtered_df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='state', y='production', data=df)
sns.stripplot(x='state', y='production', data=df)

plt.xticks(rotation=90)
plt.title("Without Jitter & Alpha")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='state', y='production', data=df)
sns.stripplot(x='state', y='production', data=df,
              jitter=True, alpha=0.4, color='red')

plt.xticks(rotation=90)
plt.title("With Jitter & Alpha")
plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.boxplot(x='state', y='production', data=df, showfliers=False)
sns.stripplot(x='state', y='production', data=df,
              jitter=True, alpha=0.5, color='black', size=3)

plt.xticks(rotation=90)
plt.title("Clean Visualization (Recommended)")
plt.show()
```

```python
import numpy as np

corr = df.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10,8))

sns.heatmap(corr,
            mask=mask,
            annot=True,
            cmap='coolwarm')

plt.show()
```

```python
corr = df.corr(numeric_only=True)

for col in corr.columns:
    for row in corr.index:
        if abs(corr.loc[row, col]) > 0.8 and row != col:
            print(row, col, corr.loc[row, col])
```

```python
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(abs(upper[column]) > 0.8)]

df_reduced = df.drop(columns=to_drop)
```
