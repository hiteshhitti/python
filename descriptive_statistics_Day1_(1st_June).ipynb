```python
# Measures of Central Tendency :

# Measure of central tendency estimate the center, or average of a dataset.
# The mean, median, mode are three ways of finding the average.

# Mean
# It is the most common way known to find the average

# Ex: 15,12,3,0,24,3
# Mean = Sum all values / Total no. of values = 57/6 = 9.5

# Median
# It is the value that is exactly in the middle of a dataset
# Arrange the data in ascending/descending order before finding median

# Ex: 0,3,3,12,15,24 (IN CASE THE NUMBER OF ENTRIES ARE EVEN)
# Middle numbers : 3 , 12
# Median : (Mean of both numbers) = 7.5

# Ex: 0,3,12,15,24 (IN CASE THE NUMBER OF ENTRIES ARE ODD)
# Middle numbers : 12
# Median : 12


# Mode
# It is simply the most popular or most frequent response value.
# A dataset can have no mode, one mode, or more than one mode.

# To find the mode, order your dataset from the lowest to highest and find the response that occurs most frequently.

# Ex: 1,3,5,12,15,24
# Mode = 0

# Ex: 0,3,3,12,15,24
# Mode = 3

# Ex: 0,3,3,12,12,15,24,24
# Mode = 3,12,24
```

```python
# Measures of Dispersion:

# It gives us an idea about the dispersion,spread and range of a data.
```

```python
# Range

# It gives us the idea of the maximum and minimum possible values and the extent in between
# Order the dataset then find the difference of the greatest and least value to find the range

# Maximum Value - Minimum Value

# Ex: 0,3,3,12,15,24
# Range : 24-0=24
```

```python
# Standard Deviation

# The standard deviation is the average amount of variability in your dataset
# It tells you on average, how far each score lies from the mean.
# The larger the standard deviation the more the variable that data/dataset is.
```

image.png

```python
# Variance

# The variance is the average of the squared deviations from the mean.
# Variance reflects the degree of spread in data/dataset.
# The more the spread in the data, the larger is in the relation to the mean.

# To find the variance:
# Simply square the std. deviation i.e. Variance = (Standard Deviation)^2
```

```python
# Inter-Quartile Range (IQR) = Q3-Q1

# Lower fence
# Q1 - 25%
# Q2 - 50%
# Q3 - 75%
# Upper fence

# Upper fence = Q3 + 1.5*(IQR)
# Lower fence = Q1 - 1.5*(IQR)

# Outliers = Upperfence - Lowerfence = [Q3 + 1.5*(IQR)] - [Q1 - 1.5*(IQR)]

# And everything below the lower fence and above the upper fence is known as outliers
# Outliers are those values that do no come inside the IQR or the box plot

# Percentiles

# 0% - 25% (Q1)
# 25% - 50% (Q2)
# 50% - 75% (Q3)

# Quartiles

# 4 quartiles are the regions made by the 3 quantiles (Q1, Q2, Q3)
```

```python
# Measure of Relationship:

# Correlation

# If any variable is having any changes observed due to change other variables then it is correlation

# Positive correlation - increase in one variable leads to increase in other variable
# Ex: Increase in speed of car will increase the chances of accident

# Negative correlation - increase in one variable leads to decrease in other variable
# Ex: Increase in speed of a car , will decrease the time to reach or time consumed

# Zero correlation - increase in one variable leads to no affect in other variable
# Ex: Increase in speed of a car , would have no affect on the price of milk
```
