```python
marks = [45, 78, 88, 32, 90]

passed = list(filter(lambda m: m >= 40, marks))
grace = list(map(lambda m: m + 5, passed))

print("Passed:", passed)
print("After Grace:", grace)
```

```python
def power(n):
    return lambda x: x ** n

square = power(2)
cube = power(3)

print(square(5))
print(cube(5))
```

```python
def is_pass(marks):
    return marks >= 40

marks = [35, 45, 60, 25, 80]

passed = list(filter(is_pass, marks))
print(passed)
```

```python
def is_even(x):
    return x % 2 == 0

def square(x):
    return x * x

nums = [1, 2, 3, 4, 5, 6]

result = list(map(square, filter(is_even, nums)))
print(result)
```
