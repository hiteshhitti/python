```python

base_url = "http://quotes.toscrape.com"

all_data = []
author_cache = {}

for page in range(1, 6):
    url = f"{base_url}/page/{page}/"
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    quotes = soup.find_all("div", class_="quote")
    
    for q in quotes:
        quote = q.find("span", class_="text").text
        author = q.find("small", class_="author").text
        
        tags = [tag.text for tag in q.find_all("a", class_="tag")]
        
        author_link = q.find("a")["href"]
        full_author_url = base_url + author_link
        
        # 🔥 Cache check
        if author in author_cache:
            about = author_cache[author]
        else:
            author_response = requests.get(full_author_url)
            author_soup = BeautifulSoup(author_response.text, "html.parser")
            
            about = author_soup.find("div", class_="author-description").text.strip()
            
            author_cache[author] = about
        
        all_data.append({
            "quote": quote,
            "author": author,
            "tags": tags,
            "about": about
        })

print(len(all_data))
```

```python
for i in range(len(all_data)):
    print(all_data[i]['quote'].strip('“”'))
    print(all_data[i]['author'].strip('“”'))
    print(all_data[i]['tags'])
    print(all_data[i]['about'].strip('“”'))
    print('*'*50)
```
