```python
n=20
for i in range(n):
    for j in range(n):
        if i == 0 or i == n -1 or j==0 or j==n-1:
            print('*', end='')
        else:
            print(' ', end='')
    print()
```

```python
num=int(input('enter number'))
if num<2:
    print('not prime')
else:
    for i in range(2,num):
        if num%i==0:
            print('not prime')
            break
    else:
        print('prime')
```

```python
text=input('enter text')
rev=''
for ch in text:
    rev=ch+rev
if text==rev:
    print('palindrome')
else:
    print('not palindrome')
```

```python
word=input()
vowels='aeiou'
vc=cc=0
for ch in word:
    if ch.isalpha():
        if ch in vowels:
            vc+=1
        else:
            cc+=1
print('vowels are: ', vc)
print('consonants are: ',cc)
```

```python
n=int(input('no of terms'))
a,b=0,1
print(a,b, end=' ')
for i in range (n):
    c=a+b
    print(c, end=' ')
    a,b=b,c
```
