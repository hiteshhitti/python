```python
queries=['date', 'time']
```

```python
query=input()
if query in queries:
    if query=='date':
        now=datetime.now()
        curr_date=now.strftime('%d-%m-%Y')
        print(curr_date)
    elif query =='time':
        now=datetime.now()
        curr_time=now.strftime('%H:%M:%S')
        print(curr_time)
else:
    print('wrong input')
```

```python
balance=5000
```

```python

card_inserted=input('instert yes/no')
if card_inserted == 'yes':
    pin=input('pin')
    if pin == '123456':
        print('1. check balance\n2. withdraw\n3. deposit')
        choice=input('choose1/2/3')
        if choice=='1':
            print('your balnce: ', balance)
        elif choice=='2':
            amount=int(input('enter amount to withdraw'))
            if amount<=balance:
                balance-=amount
                print('amount withdrawn', amount)
                print('balance now: ', balance)
            else:
                print('insufficient balance')
        elif choice=='3':
            amount=int(input('enter amount to deposit'))
            balance+=amount
            print('amount deposite: ', amount)
            print('balance now: ', balance)
        else:
            print('invalid option')
    else:
        print('wrong pin')
else:
    print('insert card')
```

```python
password=input()
if len(password)<6:
    print('password must be at lest 6 digits')
elif ' ' in password:
    print('no space allowed in password')
elif not any(i in '!@#$%^&*' for i in password):
    print('password must hve special character')
elif not any (char.isupper() for char in password):
    print('password must have upper letters')
elif not any (char.islower() for char in password):
    print('password must have small letters')
elif not any (char.isdigit() for char in password):
    print('password must have numbers')
else:
    print('password is strong')
```
