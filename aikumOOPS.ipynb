```python
class ring:
    def __init__(self, a, b , c , d):  #constructor
        self.a=a
        self.b=b
        self.c=c
        self.d=d
    def cal(self,x):
        print(self.a+self.b+x)
    def cal2(self,y):
        print(self.a+self.c*y)
    def shw(self):
        print(self.d)
```

```python
lol=ring(2,3,4,"tower")
```

```python
lol.cal(34)
```

```python
lol.cal2(56)
```

```python
lol.shw()
```

```python
class bankaccount:
    def __init__(self, name, balance=0):
        self.name=name
        self.balance=balance
        self.transactions=[]
        self.transactions.append(f"Name:-{name}")
        self.transactions.append(f"opening Balance:- {balance}")
        print("welcome to bank")
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
        else:
            print("amount must be positive")
        self.transactions.append(f"deposit: {amount} balance:- {self.balance}")
    def withdraw(self, amount):
        if amount < self.balance :
            self.balance -= amount
        else:
            print("not suficient balance")
        self.transactions.append(f"withdraw: {amount}  balance:- {self.balance}")
    def viewbalance(self):
        print(self.name, self.balance)
    def transaction(self):
        for i in self.transactions:
            print(i)
```

```python
aikum=bankaccount('aikum', 100)
```

```python
aikum.viewbalance()
```

```python
aikum.deposit(250)
```

```python
aikum.viewbalance()
```

```python
aikum.withdraw(199)
```

```python
aikum.viewbalance()
```

```python
aikum.transaction()
```

```python
from abc import ABC , abstractmethod
```

```python
class animal(ABC):
    @abstractmethod
    def sound(self):
        pass
    @abstractmethod
    def move(self):
        pass
```

```python
class zebra(animal):
    def sound(self):
        print("hehehe")
    def move(self):
        print("run")
```

```python
zebra.sound("hehehe")
```

```python
class hen(animal):
    def king(self):
        print("i am king")
```

```python
egg=hen()
```
