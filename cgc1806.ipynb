```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
def sum_rec(n):
    if n == 0:
        return 0
    return n + sum_rec(n-1)
```

```python
sum_rec(5)
```

```python
def make_discount(discount):
    def apply(price):
        return price - (price*discount)
    return apply
```

```python
summer_discount=make_discount(.20)
```

```python
summer_discount(200)
```

```python
winter_discount=make_discount(.15)
```

```python
winter_discount(200)
```

```python
def make_discount2(discount):
    def apply(price):
        return price - (price*discount)
    def apply2 (price):
        return price + (price*discount)
    return apply, apply2
```

```python
newdiscount, adddiscount=make_discount2(.20)
```

```python
newdiscount(200)
```

```python
adddiscount(200)
```

```python
def checkvowel(word):
    word=word.lower()
    vowels='aeiou'
    vc=cc=0
    for ch in word:
        if ch.isalpha():
            if ch in vowels:
                vc+=1
            else:
                cc+=1
    print('vowels are: ', vc)
    print('consonants are: ',cc)
```

```python
checkvowel('Hitesh')
```

```python
checkvowel('crocodile')
```
