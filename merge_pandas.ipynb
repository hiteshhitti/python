```python
students = pd.DataFrame({
    "roll":[1,2,3],
    "name":["Rahul","Amit","Neha"]
})
```

```python
marks = pd.DataFrame({
    "roll":[1,2,3],
    "marks":[85,78,92]
})
```

```python
df = pd.merge(students, marks, on="roll")
print(df)
```

```python
marks2 = pd.DataFrame({
    "roll":[1,2,4],
    "marks":[85,78,56]
})
```

```python
pd.merge(students, marks2, on="roll", how="outer")
```

```python
df1 = pd.DataFrame({
    "roll":[1,2,3],
    "name":["Rahul","Amit","Neha"],
    "city":["Delhi","Mumbai","Delhi"]
})
```

```python
df2 = pd.DataFrame({
    "roll":[1,2,3],
    "city":["Delhi","mohali","Pune"],
    "marks":[85,78,90]
})
```

```python
df = pd.merge(df1, df2, on=["roll","city"])
print(df)
```
