```python
#hierarchiecal agglomerative clustering
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
# Step 1: Load Dataset
df = pd.read_csv("Mall_Customers.csv")

# Step 2: Select required columns for clustering (Annual Income and Spending Score)
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values

```

```python
# Step 3: Plot Dendrogram to decide optimal number of clusters
plt.figure(figsize=(10, 6))
Z = linkage(X, method='ward')
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Customer Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Step 4: From dendrogram, let's choose k=5 (you can change this based on dendrogram)
k = 5
```

```python
# Step 5: Apply Agglomerative Clustering (No affinity parameter with ward linkage)
model = AgglomerativeClustering(n_clusters=k, linkage='ward')
y_hac = model.fit_predict(X)
```

```python
# Step 6: Visualize the Clusters
plt.figure(figsize=(8, 6))

colors = ['red', 'blue', 'green', 'cyan', 'magenta']

for i in range(k):
    plt.scatter(X[y_hac == i, 0], X[y_hac == i, 1], s=100, c=colors[i], label=f'Cluster {i+1}')

plt.title('Customer Segments using Hierarchical Agglomerative Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
