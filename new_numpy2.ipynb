```python
expenses = np.array([
    [200, 150, 300, 250, 400],   # Food
    [100, 120, 90, 110, 130],    # Travel
    [300, 400, 200, 350, 500]    # Shopping
])

print("Total Expense:", np.sum(expenses))
print("Category Wise Total:", np.sum(expenses, axis=1))
print("Day Wise Total:", np.sum(expenses, axis=0))

# Highest expense day
print("Highest Expense Day:", np.max(np.sum(expenses, axis=0)))
```

```python
marks = np.array([45, 78, 88, 32, 90, 67, 55])

print("All Marks:", marks)

# Average
avg = np.mean(marks)
print("Average Marks:", avg)

# Pass students
passed = marks[marks >= 40]
print("Passed Students:", passed)

# Topper
topper = np.max(marks)
print("Topper Marks:", topper)

# Grades
grades = np.where(marks >= 75, "A",
          np.where(marks >= 60, "B",
          np.where(marks >= 40, "C", "Fail")))

print("Grades:", grades)
```
