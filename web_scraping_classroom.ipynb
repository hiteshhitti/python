## 🔵 LESSON 1 — What is Web Scraping?

**Web Scraping** = Writing Python code that **automatically reads data from websites**

### 🌍 Real-World Uses:
- 🛒 **Price comparison** → Amazon vs Flipkart
- 📰 **News aggregator** → Collect headlines from multiple sites
- 💼 **Job listings** → Scrape jobs from Naukri, LinkedIn

### ⚙️ How It Works — 3 Steps:
```
STEP 1 → Send a request to a website  (like opening a browser)
STEP 2 → Read the HTML returned

```

> Example: `https://example.com/robots.txt`

## 🔵 LESSON 3 — Fetching a Web Page

We use **`requests.get(url)`** to download a webpage.

Think of it like: *Python opens the browser and visits the URL for you.*

### 📊 Status Codes:
| Code | Meaning |
|---|---|
| ✅ 200 | Success |
| ❌ 404 | Page Not Found |
| ❌ 403 | Forbidden (website blocked us) |
| ❌ 500 | Server Error |

````python
# ── FETCH A WEB PAGE ─────────────────────────────────
url = "https://books.toscrape.com"

response = requests.get(url)           # Send the HTTP request

print("Status Code:", response.status_code)
print("Content Type:", response.headers.get("Content-Type"))
print("\nFirst 300 characters of raw HTML:")
print(response.text[:300])
````

## 🔵 LESSON 4 — Parsing HTML with BeautifulSoup

`response.text` gives us **raw HTML** (hard to read & search).

**BeautifulSoup** converts it into a **structured object** we can search easily.

```python
soup = BeautifulSoup(response.text, "lxml")
# Now 'soup' is like a smart map of the entire HTML page
```

````python
# ── PARSE THE HTML ────────────────────────────────────
url = "https://books.toscrape.com"
response = requests.get(url)

# Create the soup object
soup = BeautifulSoup(response.text, "lxml")

# Find the page <title> tag
page_title = soup.find("title")
print("Page Title:", page_title.text)

# Pretty-print first 600 characters of formatted HTML
print("\nFormatted HTML preview:")
print(soup.prettify()[:600])
````

````python
# ── KEY DIFFERENCE ───────────────────────────────────
# soup.find()       → finds the FIRST match  (returns 1 element)
# soup.find_all()   → finds ALL matches      (returns a list)

first_h1 = soup.find("h1")             # First H1 tag only
all_h2   = soup.find_all("h2")         # Every H2 tag

print("First H1:", first_h1.text if first_h1 else "Not found")
print("Total H2 tags:", len(all_h2))
````

## 🔵 LESSON 5 — Finding Elements (4 Methods)

This is the **most important lesson** — knowing how to find the right element!

| Method | Code | Use When |
|---|---|---|
| By tag | `soup.find("h1")` | You know the HTML tag |
| By class | `soup.find("p", class_="price")` | You know the CSS class |
| Find all | `soup.find_all("article")` | You want multiple elements |
| CSS selector | `soup.select("div.card h2")` | Complex searches |

> 💡 **Tip:** Right-click any element on a website → **Inspect** → See the tag and class name!

````python
# ── METHOD A: Find by Tag Name ────────────────────────
first_h1 = soup.find("h1")
print("First H1 tag:", first_h1.text if first_h1 else "Not found")
````

````python
# ── METHOD B: Find by CSS Class ──────────────────────
# Right-click → Inspect → look for class="..."

first_book = soup.find("article", class_="product_pod")
print("First book element found:", bool(first_book))
print("\nHTML of first book card:")
print(first_book.prettify()[:400])
````

````python
# ── METHOD C: find_all() — Get Every Match ───────────
all_books = soup.find_all("article", class_="product_pod")
print("Total books on this page:", len(all_books))
````

````python
# ── METHOD D: CSS Selector (most powerful) ───────────
# Works just like CSS selectors in your stylesheet!

titles = soup.select("article.product_pod h3 a")
print("First 5 book titles:")
for title in titles[:5]:
    print("  →", title.text.strip())
````

````python
# 🔴 CLASSROOM EXERCISE 2
# 1. Open https://books.toscrape.com in your browser
# 2. Right-click a book PRICE → Inspect Element
# 3. Notice the tag is <p> and class is "price_color"
# 4. Use soup.find() to get that price

# Your code here:
price = soup.find("p", class_="price_color")
print("First book price:", price.text.strip())

# BONUS: Get ALL prices on the page
all_prices = soup.find_all("p", class_="price_color")
print("\nAll prices:")
for p in all_prices[:5]:
    print(" ", p.text.strip())
````

## 🔵 LESSON 6 — Extracting Links

Every link in HTML looks like:
```html
<a href="/catalogue/book-1.html">Book Title</a>
```

We use `tag["href"]` or `tag.get("href")` to get the URL.

### ⚠️ Relative vs Absolute URLs
| Type | Example |
|---|---|
| Relative | `catalogue/book-1.html` |
| Absolute | `https://books.toscrape.com/catalogue/book-1.html` |

Use `urljoin()` to convert relative → absolute!

````python
# ── EXTRACT ALL LINKS ────────────────────────────────
all_links = soup.find_all("a")         # Every <a> tag
print(f"Total links found: {len(all_links)}")

print("\nFirst 6 links:")
for link in all_links[:6]:
    href = link.get("href", "No href")  # .get() is safe (no error if missing)
    text = link.text.strip()
    print(f"  Text: {text[:30]:<30}  URL: {href}")
````

````python
# ── FIX RELATIVE URLS WITH urljoin ───────────────────
base_url = "https://books.toscrape.com"

# Example relative URL from the page
relative = "catalogue/a-light-in-the-attic_1000/index.html"

# Convert to absolute
absolute = urljoin(base_url, relative)

print("Relative URL:", relative)
print("Absolute URL:", absolute)

# Now get all internal links (properly fixed)
print("\nAll fixed internal links (first 5):")
for link in all_links[:5]:
    href = link.get("href", "")
    if href and not href.startswith(("http", "mailto", "tel", "#")):
        full_url = urljoin(base_url, href)
        print(" ", full_url)
````

## 🔵 LESSON 7 — Extracting Text & Data

This is where the **real scraping happens** — looping through elements and collecting data into a list of dictionaries.

```python
tag.text           # Get text (may have extra spaces)
tag.text.strip()   # Get text, remove extra spaces ✅
tag["href"]        # Get attribute value
tag.get("href")    # Safe version (no error if attribute missing)
```

````python
# ── EXTRACT BOOK DATA ────────────────────────────────
books_data = []                        # Empty list to store all books

all_book_cards = soup.find_all("article", class_="product_pod")

for book in all_book_cards:

    # ── Title (stored in the 'title' attribute of <a>) ──
    title_tag = book.find("h3").find("a")
    title     = title_tag["title"]

    # ── Price ──────────────────────────────────────────
    price = book.find("p", class_="price_color").text.strip()

    # ── Rating (stored as word in class: "star-rating Three") ──
    rating_tag = book.find("p", class_="star-rating")
    rating     = rating_tag["class"][1]  # ["star-rating", "Three"] → "Three"

    # ── Availability ───────────────────────────────────
    availability = book.find("p", class_="instock").text.strip()

    # ── Store in dictionary ────────────────────────────
    books_data.append({
        "title":        title,
        "price":        price,
        "rating":       rating,
        "availability": availability,
    })

print(f"✅ Total books scraped: {len(books_data)}")
print("\nFirst 3 books:")
for book in books_data[:3]:
    print(f"\n  Title : {book['title'][:50]}")
    print(f"  Price : {book['price']}")
    print(f"  Rating: {book['rating']} stars")
    print(f"  Stock : {book['availability']}")
````

````python
# 🔴 CLASSROOM EXERCISE 3
# Add a new field: the link to each book's detail page
#
# HINT:
#   link_tag = book.find("h3").find("a")
#   book_url = urljoin(base_url, "catalogue/" + link_tag["href"])

base_url   = "https://books.toscrape.com"
books_data2 = []

for book in all_book_cards:
    title_tag = book.find("h3").find("a")
    title     = title_tag["title"]
    price     = book.find("p", class_="price_color").text.strip()

    # ADD YOUR CODE HERE — get the book link
    book_url = urljoin(base_url, "catalogue/" + title_tag["href"])

    books_data2.append({
        "title": title,
        "price": price,
        "url":   book_url,   # ← new field!
    })

print("First 3 books with links:")
for b in books_data2[:3]:
    print(f"  {b['title'][:40]:<40}  {b['url']}")
````

## 🔵 LESSON 8 — Saving Data (JSON & CSV)

After scraping, we need to **save the data** to a file.

| Format | Best For | Opens In |
|---|---|---|
| **JSON** | Structured/nested data | Python, browsers, APIs |
| **CSV** | Simple table data | Excel, Google Sheets |

````python
# ── SAVE AS JSON ─────────────────────────────────────
json_file = "books.json"

with open(json_file, "w", encoding="utf-8") as f:
    json.dump(books_data, f, indent=4, ensure_ascii=False)

print(f"✅ Saved {len(books_data)} books to '{json_file}'")

# Read it back and display first item
with open(json_file, "r", encoding="utf-8") as f:
    loaded = json.load(f)

print("\nFirst record from JSON:")
print(json.dumps(loaded[0], indent=4))
````

````python
# ── SAVE AS CSV ──────────────────────────────────────
csv_file = "books.csv"

with open(csv_file, "w", newline="", encoding="utf-8") as f:
    fieldnames = ["title", "price", "rating", "availability"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    writer.writeheader()           # Writes column names row
    writer.writerows(books_data)   # Writes all data rows

print(f"✅ Saved {len(books_data)} books to '{csv_file}'")
print("   👉 Open books.csv in Excel to see it as a table!")
````

## 🔵 LESSON 9 — Polite Scraping (Adding Delays)

### ⚠️ Why Add Delays?

If you send **100 requests per second**, the website can:
- 🚫 **Block your IP address**
- 💥 **Crash their server** (can be illegal!)
- 🔒 **Rate-limit your requests**

**Solution:** Use `time.sleep()` to wait between requests — be a **polite scraper!**

```python
time.sleep(1.5)   # Wait 1.5 seconds before next request
```

````python
# ── SCRAPE MULTIPLE PAGES WITH DELAY ─────────────────
# books.toscrape.com has 50 pages
# Let's scrape the first 3 pages as a demo

all_pages_data = []

for page_number in range(1, 4):        # Pages 1, 2, 3

    page_url = f"https://books.toscrape.com/catalogue/page-{page_number}.html"
    print(f"\n📄 Scraping page {page_number}: {page_url}")

    page_response = requests.get(page_url)

    if page_response.status_code == 200:
        page_soup  = BeautifulSoup(page_response.text, "lxml")
        page_books = page_soup.find_all("article", class_="product_pod")

        for book in page_books:
            title = book.find("h3").find("a")["title"]
            price = book.find("p", class_="price_color").text.strip()
            all_pages_data.append({
                "title": title,
                "price": price,
                "page":  page_number
            })

        print(f"   ✅ {len(page_books)} books scraped")
    else:
        print(f"   ❌ Failed with status: {page_response.status_code}")

    print("   ⏳ Waiting 1.5 seconds...")
    time.sleep(1.5)                    # ← POLITE DELAY

print(f"\n✅ Total books from 3 pages: {len(all_pages_data)}")
````

````python
# 🔴 CLASSROOM EXERCISE 4
# Change range(1, 4) to range(1, 6) to scrape 5 pages
# Then save all_pages_data to "all_books.csv"

# Save to CSV
with open("all_books.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["title", "price", "page"])
    writer.writeheader()
    writer.writerows(all_pages_data)

print(f"✅ Saved {len(all_pages_data)} books to all_books.csv")
````

## 🔵 LESSON 10 — Mini Project: Quotes Scraper 🎓

**Target:** https://quotes.toscrape.com  
**Goal:** Scrape quotes, authors, and tags → save to JSON

### HTML Structure (from Inspect Element):
```html
<div class="quote">
    <span class="text">"The world as we have created it..."</span>
    <small class="author">Albert Einstein</small>
    <a class="tag">change</a>
    <a class="tag">deep-thoughts</a>
</div>
```

> 🎯 **Build this together in class — step by step!**

````python
# ── STEP 1: Fetch the page ────────────────────────────
quotes_url      = "https://quotes.toscrape.com"
quotes_response = requests.get(quotes_url)
quotes_soup     = BeautifulSoup(quotes_response.text, "lxml")

print("Status:", quotes_response.status_code)
print("Page title:", quotes_soup.find("title").text)
````

````python
# ── STEP 2: Find all quote blocks ────────────────────
quote_blocks = quotes_soup.find_all("div", class_="quote")
print(f"Found {len(quote_blocks)} quotes on page 1")
````

````python
# ── STEP 3: Extract data from each quote ─────────────
quotes_data = []

for block in quote_blocks:

    # Quote text
    text = block.find("span", class_="text").text.strip()

    # Author name
    author = block.find("small", class_="author").text.strip()

    # Tags (multiple!) — list comprehension
    tag_elements = block.find_all("a", class_="tag")
    tags = [tag.text.strip() for tag in tag_elements]

    quotes_data.append({
        "quote":  text,
        "author": author,
        "tags":   tags,
    })

# Display results
print(f"✅ Scraped {len(quotes_data)} quotes\n")
for q in quotes_data[:3]:
    print(f"Quote  : {q['quote'][:70]}...")
    print(f"Author : {q['author']}")
    print(f"Tags   : {', '.join(q['tags'])}")
    print()
````

````python
# ── STEP 4: Save to JSON ─────────────────────────────
with open("quotes.json", "w", encoding="utf-8") as f:
    json.dump(quotes_data, f, indent=4, ensure_ascii=False)

print("✅ quotes.json saved!")
print("\nSample from JSON:")
print(json.dumps(quotes_data[0], indent=4))
````

````python
# 🏠 HOMEWORK: Scrape ALL 10 pages
# HINTS:
#   - Page URLs: https://quotes.toscrape.com/page/1/ , /page/2/ , ...
#   - Use a for loop: for page in range(1, 11)
#   - Add time.sleep(1) between pages
#   - Collect all quotes into one list
#   - BONUS: Find the most common tag across all quotes

# Your code here:
all_quotes = []

for page in range(1, 11):              # Pages 1 to 10
    page_url = f"https://quotes.toscrape.com/page/{page}/"
    print(f"Scraping page {page}...", end=" ")

    res  = requests.get(page_url)
    psoup = BeautifulSoup(res.text, "lxml")

    blocks = psoup.find_all("div", class_="quote")
    for block in blocks:
        text   = block.find("span", class_="text").text.strip()
        author = block.find("small", class_="author").text.strip()
        tags   = [t.text.strip() for t in block.find_all("a", class_="tag")]
        all_quotes.append({"quote": text, "author": author, "tags": tags})

    print(f"✅ {len(blocks)} quotes")
    time.sleep(1)

print(f"\n✅ Total quotes scraped: {len(all_quotes)}")

# BONUS: Find the most common tag
from collections import Counter
all_tags = [tag for q in all_quotes for tag in q["tags"]]
top_tags = Counter(all_tags).most_common(5)
print("\n🏷️  Top 5 most common tags:")
for tag, count in top_tags:
    print(f"  {tag:<20} → {count} times")
````

## 📋 Quick Reference Cheat Sheet

| Code | What it does |
|---|---|
| `requests.get(url)` | Download a webpage |
| `response.status_code` | Check if request succeeded (200 = OK) |
| `response.text` | Get raw HTML as string |
| `BeautifulSoup(html, "lxml")` | Parse HTML into searchable object |
| `soup.find("tag")` | Find FIRST matching tag |
| `soup.find("tag", class_="name")` | Find by tag + CSS class |
| `soup.find_all("tag")` | Find ALL matching tags (returns list) |
| `soup.select("div.card h2")` | CSS selector (powerful!) |
| `tag.text.strip()` | Get visible text, clean whitespace |
| `tag["href"]` | Get attribute value |
| `tag.get("href", "")` | Safe attribute get (no error if missing) |
| `urljoin(base, relative)` | Convert relative URL to absolute |
| `time.sleep(1.5)` | Wait 1.5 seconds (polite scraping!) |
| `json.dump(data, file, indent=4)` | Save to JSON file |
| `json.load(file)` | Read from JSON file |
| `csv.DictWriter(file, fieldnames)` | Write list of dicts to CSV |
