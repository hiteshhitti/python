```python

import pandas as pd

# Load Dataset
df = pd.read_csv("large_messy_dataset_option2.csv")
df.head()
```

```python

# Rename columns: lowercase and replace spaces with underscore
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python

# Convert numeric-like columns safely
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

df.dtypes
```

```python

# Attempt to convert any column containing 'date' in name
for col in df.columns:
    if 'date' in col:
        df[col] = pd.to_datetime(df[col], format='mixed', dayfirst=True, errors='coerce')

df.head()
```

```python

# Clean object/string columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

df.head()
```

```python

# Example apply function (if salary column exists)
if 'salary' in df.columns:
    df['bonus_10_percent'] = df['salary'].apply(lambda x: x * 0.10 if pd.notnull(x) else x)

df.head()
```

```python

# Check missing values
df.isna().sum()
```

```python

# Save cleaned dataset
df.to_csv("cleaned_large_messy_dataset.csv", index=False)
```
