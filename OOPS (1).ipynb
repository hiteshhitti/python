Encapsulation

```python
# Wrapping all your data into a single unit is known as encapsulation.
```

```python
class MyClass:
    def __init__(self, variable):  # Constructor
        self.variable = variable

    def my_method(self):  # Method
        print("Variable value:", self.variable)
obj = MyClass("Hello, world!")
obj.my_method()
```

```python
class BankAccount:
  def __init__(self, balance): #constructor
    self.balance = balance

  def my_method(self): #method
    print("balance",self.balance)

obj = BankAccount("one,lakh")
obj.my_method()
```

Inheritance

```python
A child class is inheriting the properties of his parent class.
```

```python
#single inheritance-->one parent,one child
```

```python
class parent:
    def fun1(self):
        print("this is my parent class")
class child(parent):
    def fun2(self):
        print("this is my child class")
```

```python
c=child()
```

```python
c.fun2()
```

```python
c.fun1()
```

```python
class f:
    def fun1(self):
        print("Father class")
class m:
    def fun2(self):
        print("Mother Class")
class c(f,m):
    def fun3(self):
        print("Child class")
```

```python
oc=c()
```

```python
oc.fun3()
```

```python
oc.fun2()
```

```python
oc.fun1()
```

```python
of=f()
```

```python
of.fun1()
```

```python
of.fun2()
```

```python
of.fun3()
```

```python
# Multilevel inheritance--> grandparent parent and child relationship
```

```python
class g:
    def fun1(self):
        print('grandparent')
class p(g):
    def fun2(self):
        print('parent')
class c(p):
    def fun3(self):
        print('child')
obb=c()
```

```python
obb.fun3()
```

```python
obb.fun2()
```

```python
obb.fun1()
```

```python
class G:
    def fun1(self):
        print("Grandparent class")
class P(G):
    def fun2(self):
        print("Parent class")
class c(P):
    def fun3(self):
        print("child class")
oc=c()
```

```python
oc.fun1()
```

```python
oc.fun2()
```

```python
oc.fun3()
```

```python
#hierarchical inheritance--> one parent and multiple childs
```

```python
class parent:
    def fun(self):
        print("This is my parent class")
class c1(parent):  
    def fun1(self):
        print("This is my child1")
class c2(parent):  
    def fun2(self):
        print("This is my child2")
```

```python
obc2=c2()
```

```python
obc2.fun2()
```

```python
obc2.fun()
```

```python
obc2.fun1()
```

```python
class animal:
    def sound(self):
        pass
class cat(animal):
    def sound(self):
        print("meow")
class dog(animal):
    def sound(self):
        print("bark")

def animal_sound(a):
    a.sound()
```
