```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
honey=pd.read_csv('US_honey_dataset_updated.csv')
```

```python
twostates=honey[honey['state'].isin(['Alabama', 'Arizona'])]
```

```python
twostates=twostates[['state', 'production', 'year']]
```

```python
pivot_twostates=twostates.pivot(index='year', columns='state', values='production')
```

```python
pivot_twostates
```

```python
pivot_twostates.plot(kind='bar', figsize=(10,5))
plt.title('yearly production data')
plt.xlabel('years')
plt.ylabel('production')
plt.legend(title='state')
plt.show()
```

```python
pivot_twostates.plot(kind='line', figsize=(10,5), marker='o')
plt.title('yearly production data')
plt.xlabel('years')
plt.ylabel('production')
plt.legend(title='state')
plt.show()
```

```python
pivot_twostates.plot(kind='bar', stacked=True, figsize=(10,5))
plt.title('yearly production data')
plt.xlabel('years')
plt.ylabel('production')
plt.legend(title='state')
plt.show()
```

```python
pivot_twostates.plot(kind='barh', stacked=True, figsize=(10,5))
plt.title('yearly production data')
plt.xlabel('years')
plt.ylabel('production')
plt.legend(title='state')
plt.show()
```

```python
pivot_twostates.loc[2000].plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title('pie chart')
plt.ylabel('')
plt.show()
```
