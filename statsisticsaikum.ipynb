```python
plt.figure(figsize=(10,5))
sn.barplot(x=ourdata3["state"], y=ourdata3["colonies_number"])
plt.xlabel("states")
plt.ylabel("colonies")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(ourdata3["state"], ourdata3["colonies_number"])
plt.xlabel("states")
plt.ylabel("colonies")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
chart=sn.barplot(x="state", y="colonies_number", data=ourdata3, palette="magma")
plt.title("data")
plt.xlabel("states")
plt.ylabel("colonies")
plt.xticks(rotation=45)
plt.show()
```
