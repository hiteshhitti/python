```python
data = np.array([
    [30, 10, 10],   # Online
    [15, 25, 10],   # Offline
    [35,  5, 10]    # Hybrid
])
```

```python
chi2, p, dof, expected = stats.chi2_contingency(data)

print("Chi-square value:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected values:\n", expected)
```

```python
if p < alpha:
    print("Reject H0 → Relation exists")
else:
    print("Fail to Reject H0 → No strong relation")
```

```python
# Observed frequency table
observed = np.array([
    [20, 15, 15],   # School
    [30, 10, 10],   # Graduate
    [25, 20, 5]     # Postgraduate
])
```

```python
chi2, p, dof, expected = chi2_contingency(observed)
```

```python
print("Chi-Square Value:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

```python
n = observed.sum()
k = min(observed.shape)  # smaller of rows or columns

cramers_v = np.sqrt(chi2 / (n * (k - 1)))

print("Cramer's V:", cramers_v)
```

```python
group_A = [65, 75, 60, 80]
group_B = [80, 75, 85, 92]
group_C = [68, 65, 70, 73]
```

```python
f_stat, p_value = f_oneway(group_A, group_B, group_C)

print("F-value:", f_stat)
print("p-value:", p_value)
```

```python
data = group_A + group_B + group_C
groups = (["A"]*4 + ["B"]*4 + ["C"]*4)

df = pd.DataFrame({
    "score": data,
    "group": groups
})
```

```python
tukey = pairwise_tukeyhsd(endog=df["score"],
                          groups=df["group"],
                          alpha=0.05)

print(tukey)
```
