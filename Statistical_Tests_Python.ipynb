## 1️⃣ One Sample T-Test (Manual + CI)

```python

import numpy as np
from scipy import stats

data = [22, 25, 19, 24, 23, 27, 21, 26]
mu_0 = 20

n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)

t_stat = (mean - mu_0) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))

confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_error = t_critical * (std / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

print("Sample Mean:", mean)
print("T-statistic:", t_stat)
print("P-value:", p_value)
print("95% Confidence Interval:", (lower, upper))
print("t critical:- ",t_critical)
print("margin error:- ",margin_error)
```

```python
t_stat, p_value = stats.ttest_1samp(data, mu_0)
print(t_stat, p_value)
if p_value < 0.05:
    print("Conclusion: Significant Difference")
else:
    print("Conclusion: No Significant Difference")
```

## 2️⃣ Independent T-Test

```python

group1 = [10, 12, 14, 16, 18]
group2 = [8, 9, 11, 13, 15]

t_stat, p_value = stats.ttest_ind(group1, group2)

print("T-stat:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Conclusion: Significant Difference")
else:
    print("Conclusion: No Significant Difference")
```

```python
import numpy as np
from scipy import stats

group1 = np.array([10, 12, 14, 16, 18])
group2 = np.array([8, 9, 11, 13, 15])

n1 = len(group1)
n2 = len(group2)

mean1 = np.mean(group1)
mean2 = np.mean(group2)

std1 = np.std(group1, ddof=1)
std2 = np.std(group2, ddof=1)

# Pooled SD
sp = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

se = sp * np.sqrt(1/n1 + 1/n2)

t_stat = (mean1 - mean2) / se
df = n1 + n2 - 2
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=df))

confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=df)
margin_error = t_critical * se

lower = (mean1 - mean2) - margin_error
upper = (mean1 - mean2) + margin_error

# -------- Clean Output --------

print("Mean Difference:", round(mean1 - mean2, 2))
print("T-statistic:", round(t_stat, 3))
print("P-value:", round(p_value, 3))
print("95% Confidence Interval:", (round(lower,2), round(upper,2)))

if p_value < 0.05:
    print("Conclusion: Significant Difference")
else:
    print("Conclusion: No Significant Difference")
```

## 3️⃣ Paired T-Test

```python

before = [10, 12, 14, 16, 18]
after  = [12, 14, 15, 18, 20]

t_stat, p_value = stats.ttest_rel(before, after)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

## 4️⃣ One-Way ANOVA

```python

group1 = [10, 12, 14]
group2 = [15, 17, 19]
group3 = [20, 22, 24]

f_stat, p_value = stats.f_oneway(group1, group2, group3)

print("F-stat:", f_stat)
print("P-value:", p_value)
```

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Data
groupA = [10, 12, 14]
groupB = [15, 17, 19]
groupC = [20, 22, 24]

# Step 1: ANOVA
f_stat, p_value = stats.f_oneway(groupA, groupB, groupC)
print("ANOVA P-value:", p_value)

# Step 2: Tukey Test (Post Hoc)

# Combine data
data = groupA + groupB + groupC
groups = ['A']*3 + ['B']*3 + ['C']*3

df = pd.DataFrame({
    'values': data,
    'group': groups
})

tukey = pairwise_tukeyhsd(endog=df['values'],
                          groups=df['group'],
                          alpha=0.05)

print(tukey)
```

## 5️⃣ Pearson Correlation

```python

x = [1,2,3,4,5]
y = [2,5,7,10,11]

corr, p_value = stats.pearsonr(x, y)

print("Correlation:", corr)
print("P-value:", p_value)
```

## 6️⃣ Chi-Square Test

```python

from scipy.stats import chi2_contingency

table = np.array([[10, 20, 30],
         [6,  9,  17]])

n = table.sum()
k = min(table.shape)

chi2, p, dof, expected = chi2_contingency(table)
cramers_v = np.sqrt(chi2 / (n * (k - 1)))

print("Chi-square:", chi2)
print("P-value:", p)
print("DOF:", dof)
print("Expected Frequencies:\n", expected)
print("Cramer's V:", cramers_v)


if p < 0.05:
    print("\nResult: There is a statistically significant relationship between the groups.")
else:
    print("\nResult: There is NO statistically significant relationship between the groups.")

# Strength interpretation
if cramers_v < 0.1:
    strength = "Very Weak"
elif cramers_v < 0.3:
    strength = "Weak"
elif cramers_v < 0.5:
    strength = "Moderate"
else:
    strength = "Strong"

print("Strength of Association:", strength)
```

## 7️⃣ Linear Regression

```python

from scipy import stats

x = [1,2,3,4,5]
y = [2,4,5,4,5]

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
x_new = 6
y_pred = intercept + slope * x_new

# print("Slope:", slope)
# print("Intercept:", intercept)
print("Predicted y for x=6:", round(y_pred, 3))
```

## 8️⃣ Outliers using Z-Score

```python

data = [10, 12, 14, 15, 100]

z_scores = np.abs(stats.zscore(data))
outliers = [data[i] for i in range(len(data)) if z_scores[i] > 3]

print("Outliers (Z-score method):", outliers)
```

## 9️⃣ Outliers using IQR

```python

data = [10, 12, 14, 15, 100]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers (IQR method):", outliers)
```

```python
#outliers using modified z score


import numpy as np

data = np.array([10, 12, 14, 15, 100])

# Step 1: Median
median = np.median(data)

# Step 2: MAD (Median Absolute Deviation)
mad = np.median(np.abs(data - median))

# Step 3: Modified Z-Score
modified_z_scores = 0.6745 * (data - median) / mad

# Step 4: Detect Outliers
outliers = data[np.abs(modified_z_scores) > 3.5]

print("Median:", median)
print("MAD:", mad)
print("Modified Z-Scores:", modified_z_scores)
print("Outliers:", outliers)
```
